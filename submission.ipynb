```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("Train_dataset.xlsx")
```

```python
df.columns
```

```python
possible_factors=['people_ID','Region','Gender','Married','Children','Mode_transport','cases/1M','Deaths/1M','comorbidity','Age','Coma score', 'Pulmonary score',
       'cardiological pressure', 'Diuresis', 'Platelets', 'HBB', 'd-dimer',
       'Heart rate', 'HDL cholesterol', 'Charlson Index', 'Blood Glucose',
       'Insurance', 'salary', 'FT/month']
```

```python
xtrain=df[possible_factors]
ytrain=df['Class']
```

```python
import sklearn
```

```python
from sklearn.feature_selection import SelectKBest
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
for i in range(len(xtrain['Pulmonary score'])):
    xtrain['Pulmonary score'][i]=xtrain['Pulmonary score'][i][1:]
```

```python
xtrain['Mode_transport'].fillna(xtrain['Mode_transport'].mode()[0], inplace=True)
xtrain['comorbidity'].fillna(xtrain['comorbidity'].mode()[0], inplace=True)
xtrain['Children'].fillna(xtrain['Children'].mode()[0], inplace=True)
xtrain['cardiological pressure'].fillna(xtrain['cardiological pressure'].mode()[0], inplace=True)
xtrain['Diuresis'].fillna(xtrain['Diuresis'].mode()[0], inplace=True)
xtrain['Platelets'].fillna(xtrain['Platelets'].mode()[0], inplace=True)
xtrain['HBB'].fillna(xtrain['HBB'].mode()[0], inplace=True)
xtrain['d-dimer'].fillna(xtrain['d-dimer'].mode()[0], inplace=True)
xtrain['Heart rate'].fillna(xtrain['Heart rate'].mode()[0], inplace=True)
xtrain['HDL cholesterol'].fillna(xtrain['HDL cholesterol'].mode()[0], inplace=True)
xtrain['Insurance'].fillna(xtrain['Insurance'].mode()[0], inplace=True)
xtrain['FT/month'].fillna(xtrain['FT/month'].mode()[0], inplace=True)
xtrain.head()
```

```python
xtrain.set_index('people_ID')
```

```python
from sklearn.compose import ColumnTransformer
columnTransformer = ColumnTransformer([('encoder', OneHotEncoder(), [1,2,3,5,8,12])], remainder='passthrough')
xtrain = pd.DataFrame(columnTransformer.fit_transform(xtrain), dtype = np.str)
```

```python
from sklearn.feature_selection import chi2,f_classif
```

```python
bestfeatures = SelectKBest(f_classif,k='all')
fit=bestfeatures.fit(xtrain,ytrain)
dfscores = pd.DataFrame(fit.scores_)
dfcolumns = pd.DataFrame(xtrain.columns)
dfpvalues=pd.DataFrame(fit.pvalues_)
ytrain=pd.DataFrame(ytrain)
featureScores = pd.concat([dfcolumns,dfscores,dfpvalues],axis=1)
featureScores.columns = ['Specs','Score','Pvalues']
Xtrain=xtrain[featureScores[featureScores['Pvalues']<.3]['Specs']]
#print(featureScores.sort_values('Score'))
Xtrain.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
train_X, val_X, train_y, val_y = train_test_split(Xtrain,ytrain, test_size = 0.25, random_state = 14)
from sklearn.ensemble import RandomForestClassifier

# Create the model with 100 trees
model = RandomForestClassifier(n_estimators=100, 
                               bootstrap = True,
                               max_features = 'sqrt')
# Fit on training data
model.fit(train_X,train_y)

# Get the mean absolute error on the validation data
predicted_prices = model.predict(val_X)
MAE = mean_absolute_error(val_y , predicted_prices)
acc =accuracy_score(val_y , predicted_prices)
print('Random forest validation acc = ', acc)
```
